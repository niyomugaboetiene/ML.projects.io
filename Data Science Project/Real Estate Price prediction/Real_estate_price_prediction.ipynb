```python
# drop null values
df3 = df2.dropna()
df3.isna().sum()
```

```python
df3["BHK"] = df3["size"].apply(lambda x: int(x.split(" ")[0]))
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(20)
```

```python
import re

def convert_sqft_in_num(x):
    token = str(x).split('-')
    
    if len(token) == 2:
        try:
            return (float(token[0]) + float(token[1])) / 2
        except:
            return None
    
    match = re.search(r'\d+\.?\d*', str(x))
    if match:
        return float(match.group())
    
    return None
```

```python
print(convert_sqft_in_num('2100 - 2850'))
```

```python
print(convert_sqft_in_num('34.46Sq. Meter'))
```

```python
df4 = df3.copy()

df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_in_num)
df4.head()
```

```python
df4.iloc[30]
```

```python
(2100 + 2850) / 2
```

```python
df5 = df4.copy()

df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
len(df5['location'].unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip())

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_less_than_10_stats = location_stats[location_stats<=10]
location_less_than_10_stats
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_less_than_10_stats else x)
len(df5.location.unique())
```

```python
df5.head(10)
```

```python
# Outlier removal
df5[df5.total_sqft / df5.BHK < 300].head()
```

```python
df5.shape
```

```python
df6 = df5[~(df5.total_sqft / df5.BHK < 300)]
```

```python
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python

def remove_pps_outlier(df):
    df_out = pd.DataFrame()
    
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        std = np.std(subdf.price_per_sqft)
        
        reduced_df = subdf[
            (subdf.price_per_sqft > (m - std)) & 
            (subdf.price_per_sqft <= (m + std))
        ]
        
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    
    return df_out

df7 = remove_pps_outlier(df6)
df7.shape
```

```python
def plot_scatter_chart(df, location):
    bhk2 = df[(df.location == location) & (df.BHK == 2)]
    bhk3 = df[(df.location == location) & (df.BHK == 3)]

    plt.scatter(bhk2.total_sqft, bhk2.price, color='blue', marker='.', s=50, label='BHK 2')
    plt.scatter(bhk3.total_sqft, bhk3.price, color='green', marker='+', s=50, label='BHK 3')
    plt.xlabel('Total square feet')
    plt.ylabel('Price')
    plt.title(location)
    plt.legend()
    plt.show()

plot_scatter_chart(df7, 'Rajaji Nagal')
```
